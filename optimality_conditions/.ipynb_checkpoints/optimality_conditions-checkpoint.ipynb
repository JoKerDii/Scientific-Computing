## 1. First-order optimality condition <a class="anchor" id="first-bullet"></a>


If $f: \mathbb{R}^n \rightarrow \mathbb{R}$ is differentiable, then the **gradient vector** $\nabla f : \mathbb{R}^n \rightarrow \mathbb{R}^n$ is

$$
\nabla f(x) = \begin{bmatrix}{\partial f(x) \over \partial x_1} \\ {\partial f(x) \over \partial x_2} \\ \vdots \\{\partial f(x) \over \partial x_n} \end{bmatrix}
$$

where $ - \nabla f(x)$ is the steepest descent direction, from **Cauchy-Schwarz**.

This follows from Taylor's theorem for $f: \mathbb{R}^n \rightarrow \mathbb{R}$,

$$
f(x+ \delta) = f(x) + \nabla f(x)^T \delta + H.O.T
$$

Let $\delta = -\epsilon \nabla f(x)$ for $\epsilon > 0$ and suppose that $\nabla f(x) \neq 0$, then

$$
f(x- \epsilon \nabla f(x)) \approx f(x) - \epsilon \nabla f(x)^T \nabla f(x) < f(x)
$$

We see that a necessary condition for a local minimum at $x^* \in S$ is that $\nabla f(x^*) = 0$. The condition $\nabla f(x^*)=0$ is called a **first-order necessary condition for optimality**. $x^* \in S$ that satisfies the first-order optimality condition is called a **critical point** of $f$. A critical point can be a local min, local max, or saddle point.

---

## 2. Second order optimality condition <a class="anchor" id="second-bullet"></a>

If $f: \mathbb{R}^n \rightarrow \mathbb{R}$ is twice differentiable, then the **Hessian** is the matrix-valued function $H_f : \mathbb{R}^n \rightarrow \mathbb{R}^{n \times n}$.

$$
H_f(x) = \begin{bmatrix} {\partial^2 f(x) \over \partial x_1^2} & {\partial^2 f(x) \over \partial x_1x_2} & ...&{\partial^2 f(x) \over \partial x_1x_n}\\{\partial^2 f(x) \over \partial x_2x_1} & {\partial^2 f(x) \over \partial x_2^2} & ... & {\partial^2 f(x) \over \partial x_2x_n}\\ \vdots & \vdots & \ddots & \vdots \\ {\partial^2 f(x) \over \partial x_nx_1}  & {\partial^2 f(x) \over \partial x_nx_1}  & ... & {\partial^2 f(x) \over \partial x_n^2}  \end{bmatrix}
$$

If the second partial derivatives of $f$ are continuous, then $\partial^2/\partial x_i \partial x_j = \partial^2/\partial x_j \partial x_i$, and $H_f$ is symmetric.

Suppose we found a critical point $x^*$ such that $\nabla f(x^*) = 0$, from Taylor's theorem, for $\delta \in \mathbb{R}^n$, we have

$$
\begin{align}
f(x^* + \delta) &= f(x^*) + \nabla f(x^*)^T \delta + {1\over 2} \delta^T H_f(x^* + \eta \delta)\delta \\
&= f(x^*)+ {1\over 2} \delta^T H_f(x^* + \eta \delta)\delta 
\end{align}
$$

for some $\eta \in (0,1)$.

Recall that $A$ is **positive definite** if $x^TAx > 0$. Suppose $H_f(X^*)$ is positive definite, then by continuity $H_f(x^* + \eta \delta)$ is also positive definite for $||\delta||$ sufficiently small, so that

$$
\delta^T H_f(x^* + \eta \delta) \delta > 0
$$

Hence, we have $f(x^* + \delta) > f(x^*)$ for $||\delta||$ sufficiently small, i.e. $f(x^*)$ is a local minimum.

Therefore, positive definiteness of $H_f$ at a critical point $x^*$ is a **second-order sufficient condition for a local minimum**.

---

* Test for positive definiteness?
    * A matrix is positive definite iff all eigenvalues of it are positive.
    * A matrix is negative definite iff all eigenvalues of it are negative.
    * A matrix with positive and negative eigenvalues is indefinite.


* We can classify critical points as:

$$
\begin{align}
&H_f(x^*) \text{ positive definite } &\implies& x^* \text{ is a local minimum}\\
&H_f(x^*) \text{ negative definite } &\implies& x^* \text{ is a local maximum}\\
&H_f(x^*) \text{ indefinite } &\implies& x^* \text{ is a saddle point}\\
\end{align}
$$

* We can know about convexity:

    * If $H_f(x)$ is positive definite, then $f$ is convex on some convex neighborhood of $x$. 
    * If $H_f(x)$ is positive definite for all $x \in S$, where $S$ is a convex set, then $f$ is convex on $S$.

## No-constrain optimization case <a class="anchor" id="third-bullet"></a>

Consider $f(x) = 2x_1^3 + 3x_1^2 + 12 x_1 x_2 + 3 x_2^2 - 6 x_2 + 6$.

Then set

$$
\nabla f(x) = \begin{bmatrix} 6x_1^2 + 6 x_1 + 12x_2 \\ 12 x_1 + 6 x_2 - 6\end{bmatrix} = 0
$$

to find critical points $[1, -1]^T$ and $[2,-3]^T$.

The Hessian is

$$
H_f(x) = \begin{bmatrix} 12x_1 + 6 & 12 \\ 12 & 6\end{bmatrix}
$$

and for each critical point the Hessian is

$$
H_f(1,-1) = \begin{bmatrix} 18 & 12 \\ 12 & 6\end{bmatrix}, \text{ which has eigenvalues } 25.4, -1.4\\
H_f(2,-3) = \begin{bmatrix} 30 & 12 \\ 12 & 6\end{bmatrix}, \text{ which has eigenvalues } 35.0, 1.0
$$

Hence $[2,-3]^T$ is a local minimum, whereas $[1,-1]^T$ is a saddle point.

The following program shows two critical points.

```python
import numpy as np
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt

# function with critical points at (2,-3) (local min)
# and (1,-1) (saddle)
def f(x1,x2):
    return 2*x1*x1*x1+3*x1*x1+12*x1*x2+3*x2*x2-6*x2+6

def grad_f(x1,x2):
    return np.array([6*x1*x1+6*x1+12*x2,
                    12*x1+6*x2-6])

def hess_f(x1,x2):
    return np.array([[12*x1+6,12], [12,6]])

cps = np.array([[2,-3], [1,-1]])

# plotting
ax=plt.figure(figsize=(20,15)).add_subplot(projection='3d')
x1 = np.linspace(0,3,1000)
x2 = np.linspace(-4,0,1000)
X1, X2 = np.meshgrid(x1,x2)

# plot 3d surface
ax.plot_surface(X1,X2,f(X1,X2), rstride=50, cstride=50, alpha=0.5, cmap=plt.cm.gnuplot)

# plot a projection of the contours for the z dimension.
levels = np.logspace(-3,2,200)
ax.contour(X1,X2, f(X1,X2), zdir='z', offset=-50, linewidths=0.5,
          levels = levels, cmap=plt.cm.gnuplot)
ax.scatter(cps[:,0], cps[:,1], [-50,-50], color='k')

ax.set_xlim(0,3)
ax.set_ylim(-4,0)
ax.set_zlim(-50,100)

ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('f')

plt.show()
```
