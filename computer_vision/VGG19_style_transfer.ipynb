# 4. Style Transfer <a class="anchor" id="fourth-bullet"></a>

## 4.1 Paper Explained
[[paper]](https://www.cv-foundation.org/openaccess/content_cvpr_2016/papers/Gatys_Image_Style_Transfer_CVPR_2016_paper.pdf)

* The authors emphasized that the model care more about the actual content of the image compared to the detailed pixels. So they chose the feature responses in higher layers of the network as the content representation. In actual implementation, the 4_2 convolution layer of the pre-trained VGG19 network is used as content extractor.

* They used the correlations between the different filter responses over the spatial extent of the feature maps - the dot product of the same layer. In this way, a representation of the input image with its style information but not the arrangement can be obtained. These correlations between feature maps are known as gram matrix. In actual implementation, the conv1_1, 2_1,3_1,4_1,5_1 are used with varied style weights for calculating gram matrix.

* The loss function for content loss and style loss are mean square error. The total loss is the weighted (content weight alpha and style weight beta) sum of the two loss functions. The alpha and beta are called hyperparameters.

## 4.2 Implementation

* The input image was taken somewhere in Cambridge, Boston, US. The style image is a cartoon image download from internet.

* The implementation using PyTorch follows the methods described in the original paper.

```python
%matplotlib inline 
import torch
import torch.optim as optim
from torchvision import transforms,models
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Pre-trained model
vgg = models.vgg19(pretrained=True).features

# Fix training parameters
for param in vgg.parameters():
    param.requires_grad_(False)

# Move model to GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
vgg.to(device)
```

```python
def load_image(img_path, max_size=400, shape=None):
    
    image = Image.open(img_path).convert('RGB')
    if max(image.size) > max_size:
        size = max_size
    else:
        size = max(image.size)
    
    if shape is not None: 
        size = shape
 
    in_transform = transforms.Compose([
                        transforms.Resize(size),
                        transforms.ToTensor(),
                        transforms.Normalize((0.5, 0.5, 0.5), 
                                             (0.5, 0.5, 0.5))])
    image = in_transform(image).unsqueeze(0)
    return image

def im_convert(tensor):
    image = tensor.to("cpu").clone().detach()
    image = image.numpy().squeeze()
    image = image.transpose(1,2,0)
    image = image * np.array((0.5, 0.5, 0.5)) + np.array((0.5, 0.5, 0.5))
    image = image.clip(0, 1)
    return image

# Pick images
content = load_image('cambridge1.jpg').to(device)
style = load_image('youhua1.jpeg', shape=content.shape[-2:]).to(device)

# Show images
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))
ax1.imshow(im_convert(content))
ax1.axis("off")
ax2.imshow(im_convert(style))
ax2.axis("off")
```

The `get_feature` function is used to extract features from convolution layers.

* Content extraction: Only one conv layer is sufficient to extract content, because this payer is deeper into our neural network and will provide high depth image feature. This is due to the fact that the pre-trained object detection convolution on neural networks become very effective at representing content elements. The content elements are more pronounced in the deeper layers of the networks. Matching these content features with the target image features will allow for optimal content creation in the target image.

* Style extraction: The remaining five layers will be used for style extraction. Getting style of features from various layers through the network allows for the optimal style creation. Style as oppose to content is more complex to re-create, as more features will give a more complete picture of the style elements.

* See paper: "Image Style Transfer Using Convolutional Neural Networks" 
https://www.cv-foundation.org/openaccess/content_cvpr_2016/papers/Gatys_Image_Style_Transfer_CVPR_2016_paper.pdf

The `gram_matrix` function is used for more effective style feature extraction.

* Generally, any features extracted from the convolutional network still hold significant content related information such as object structure and positioning. By applying a gram matrix to the extracted features, the content information is eliminated, while the style information is maintained.

* See paper: "Texture Synthesis using Convolutional Neural Networks" https://proceedings.neurips.cc/paper/2015/file/a5e00132373a7031000fd987a3c9f87b-Paper.pdf




```python
def get_features(image, model):
    """
    Extract content and style features from conv layers.
    """
    layers = {'0': 'conv1_1',
              '5': 'conv2_1', 
              '10': 'conv3_1', 
              '19': 'conv4_1',
              '21': 'conv4_2',  # Content Extraction
              '28': 'conv5_1'}
        
    features = {}

    for name, layer in model._modules.items(): # iterate through all layers and layer names (integer number) in the model
        image = layer(image) # feed image through the neural network 
        if name in layers: # extract the output feature of the layers of interest
            features[layers[name]] = image # store the output feature of that layer to the dictionary with a key of layers string value
            
    return features

def gram_matrix(tensor):
    """
    Extract more effective style features.
    """
    _, d, h, w = tensor.size() # batch size, depth, height, width
    tensor = tensor.view(d, h * w) # reshape the tensor to be 2D
    gram = torch.mm(tensor, tensor.t()) 
    return gram
```

```python
# Content extraction and style extraction
content_features = get_features(content, vgg)
style_features = get_features(style, vgg)

# Get gram matrix for each extracted feature 
style_grams = {layer: gram_matrix(style_features[layer]) for layer in style_features}
```

Weighted layers and weighted features:

* We can choose to prioritize certain layers over other layers by associating certain weight parameters with each layer. Layers close to the beginning of the model are usually more effective at recreating style features, while later layers offer additional variety towards the style elements. So a customized arrangement which weights the early layers more heavily would be ideal.

* The parameters alpha and beta are used to customize our final target image by defining the ratio of style to content. They are referred to alpha and beta according to the style transfer paper. The lower this ratio gets the more style is implemented into the final targe image.


```python
style_weights = {'conv1_1': 1.,
                 'conv2_1': 0.75,
                 'conv3_1': 0.2,
                 'conv4_1': 0.2,
                 'conv5_1': 0.2}

content_weight = 1  # alpha: content image weight
style_weight = 1e6  # beta: style image weight
```

Define the target image
* We start with the original target image being a clone of our content image.
The optimization process is going to try and maintain the content of the image 
while applying more style from the style image with each iteration of teh optimization.
Since we want this target image to be adjusted along with the optimization process, 
thus the gradients are required.

Training steps
* The more steps we take the lower the loss value and the more effective the style transfer. We need to balance the results and time efficiency.

```python
# Define the target image
target = content.clone().requires_grad_(True).to(device)

# Get updated image every 300 iterations
show_every = 300 
optimizer = optim.Adam([target], lr=0.003)

# Training steps. 2100 is the minimum required for a decent result.
steps = 2100 

# Store the target images through out the training process.
# Define a maximum array size 300, We have 300 frames in total to visualize the optimization process.
height, width, channels = im_convert(target).shape
image_array = np.empty(shape=(300, height, width, channels)) 
capture_frame = steps/300 # capture a frame every 7 steps.
counter = 0 # track the frame index
```

```python
for ii in range(1, steps+1):
    target_features = get_features(target, vgg)

    # Minimize the mean squared error of the content feature extracted and the content of the target image. 
    # conv4_2 is used for content extraction. 
    content_loss = torch.mean((target_features['conv4_2'] - content_features['conv4_2'])**2) 
    style_loss = 0 # a combined loss from 5 different layers
  
    # Interate through 5 layers and calculate the weighted loss
    for layer in style_weights:
        target_feature = target_features[layer]
        target_gram = gram_matrix(target_feature)
        style_gram = style_grams[layer]

        # Minimize the mean squared error of the extracted style (gram matrix of style) and the style (gram matrix) of the target image.
        # The loss are weighted and aggregated.
        layer_style_loss = style_weights[layer] * torch.mean((target_gram - style_gram)**2)

        # The loss has to be normalized to get a unit loss
        _, d, h, w = target_feature.shape
        style_loss += layer_style_loss / (d * h * w)
  
    # The total loss is a sum of weighted content loss and style loss.
    total_loss = content_weight * content_loss + style_weight * style_loss

    optimizer.zero_grad()
    total_loss.backward()
    # The target image is going to take on more style from the style image while maintaining the content of the content image.
    optimizer.step() 
  
    # Visualization through the process
    if  ii % show_every == 0: # show every 300
        print('Total loss: ', total_loss.item())
        print('Iteration: ', ii)
        plt.imshow(im_convert(target))
        plt.axis("off")
        plt.show()

    if ii % capture_frame == 0:
        image_array[counter] = im_convert(target)
        counter = counter + 1
```
